```python
import numpy as np
import pandas as pd
```

```python
inst_A = pd.read_csv('data/TSPA.csv', sep=";", header=None, names=["x", "y", "cost"])
inst_B = pd.read_csv('data/TSPB.csv', sep=";", header=None, names=["x", "y", "cost"])
inst_C = pd.read_csv('data/TSPC.csv', sep=";", header=None, names=["x", "y", "cost"])
inst_D = pd.read_csv('data/TSPD.csv', sep=";", header=None, names=["x", "y", "cost"])
inst_D
```

```python
def cost(v1, v2):
    x1, y1, cost1 = v1[0], v1[1], v1[2]
    x2, y2, cost2 = v2[0], v2[1], v2[2]

    # calculate distance between two vertices
    dist = np.round(np.linalg.norm(np.array((x1, y1)) - np.array((x2, y2))), 0)
    
    # add cost of the second vertex
    return dist + cost2
```

```python
# create a matrix with all the distances and costs of choosing a certain node: 
#
# ~ = dist(v0, v1) + cost(v1)
# 
#   | 0 | 1 | 2 | 3 | ... |
# ______________________________
# 0 | x | ~ |   |   |
# 1 |   | x |   |   |
# 2 |   |   | x |   |
# ...

def dist_matrix(inst):
    matrix = []
    for i in range(len(inst)):
        dists = []
        for ii in range(len(inst)):
            # calculate the cost of choosing the vertex
            dists.append(cost(inst.loc[i], inst.loc[ii]))
        matrix.append(dists)   
    return pd.DataFrame(matrix)

dist_A = dist_matrix(inst_A)
dist_B = dist_matrix(inst_B)
dist_C = dist_matrix(inst_C)
dist_D = dist_matrix(inst_D)
```
