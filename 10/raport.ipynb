### ILS - Perturbation

As in this assignment the local search algorithm stays intact, we provide the description of the perturbation performed in the experiments instead of the whole pseudocode.

To achieve the best results, perturbation needed to be different than existing neighborhood methods.

In our case, this operation works as follows:

```
function perturb(solution):
    node_indices = get_random_indices(solution, size=2) # sample two random integers from range [0, len(solution))
    node_indices.sort() # make sure they are in increasing order

    new_solution = shuffle(solution[node_indices[0] : node_indices[1]]) # perform a random shuffle operation on elements in our chosen node_indices range

    return new_solution
```
