```
def multi_segment_shuffle_perturbation(solution):
    num_segments = randint(2, len(solution)//2)
    
    for _ in range(num_segments):
        start_idx, end_idx = sort(random.choice(len(solution), 2))
        shuffle(new_solution[start_idx:end_idx])

    return new_solution
```
